```python
allowed_users = ["MD", "MANAGER", "STAFF"]
user = input("Enter your designation")
print(user in allowed_users)
```

```python
#Ask the user to enter a fruit name and check if it exists in the list
fruits = ["apple", "banana", "lemon", "grape", "orange"]
fruit = input("Enter the fruit name")
print(fruit in fruits)
```
